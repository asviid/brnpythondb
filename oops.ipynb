OOPS (Object-Oriented Programming System) is a programming approach where we model real-world things as objects.
Objects are created from classes, which act like blueprints.


A class is a blueprint or template used to create objects.
An object is an instance of a class (a real thing created from the blueprint).

Functions inside a class are called methods.
The __init__() method is used to initialize data when an object is created.
self refers to the current object.t connects methods to object data .Without self, Python doesn’t know which object’s data to use

```python
# Define a class using the 'class' keyword and class name should be capitalized
class Car:
    # Define a method within the class using 'def' keyword   
    def start(self):   
        print("Car is starting")

c1 = Car()       # Create object
c1.start()       # Call method

```

```python
# Define a class using the 'class' keyword
class Student:

    # Constructor method - runs automatically when an object is created
    def __init__(self, name, age): # Constructor with parameters name and age
        self.name = name   # Store student's name
        self.age = age     # Store student's age

# Create an object (instance) of the Student class
s1 = Student("Riya", 15)
# Accessing attributes directly without using a method
print("Name:", s1.name)
print("Age:", s1.age)
```

```python
# Define a class using the 'class' keyword
class Student:

    # Constructor method - runs automatically when an object is created
    def __init__(self, name, age):
        self.name = name   # Store student's name
        self.age = age     # Store student's age

    # Method to display student details
    def display(self):
        print("Name:", self.name)  # Print name
        print("Age:", self.age)    # Print age


# Create an object (instance) of the Student class
s1 = Student("Riya", 15)

# Call the method using the object
s1.display()
```

<h1>Inheritance <h1>
<h4>Inheritance means one class (child) uses the properties and methods of another class (parent).<h4>

```python
#person class person can student right so here student using person class features
#A Student is a Person
# So, Student can use Person’s features.
# Parent class
class Person:
    # Constructor method
    def __init__(self, name):
        self.name = name      # Store name
    # Method to speak
    def speak(self):
        print("My name is", self.name)


# Child class inherits from Person
class Student(Person):

    def study(self):
        print(self.name, "is studying")


# Create object of Student class
s1 = Student("Amit")
# to output the name attribute from parent class
print(s1.name)  # Inherited attribute
# Call method from parent class
s1.speak()      # Inherited method

# Call method from child class
s1.study()


print("<!-- OTHER EXAMPLE-->")# Parent class
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.sound()   # From Animal
d.bark()    # From Dog
```

<h1>Polymorphism<h1>
<h5>one name , has different function<h5>
<h4>Polymorphism means the same method name, but it works differently for different objects.<h4>

```python
class Animal:

    # Constructor method i creted this to input name of your pet 
    def __init__(self , name):
        self.name = name    
    # Method to make sound so here it print petname + it makes sound 
    def sound(self): 
        print(f"{self.name} makes sound")
    # Derived class inheriting from Animal like using cocept of inheritance
class Dog(Animal):
    # Method overriding to provide specific implementation for Dog
    def sound(self):
        print(f"{self.name} barks") # print petname + it barks
class Cat(Animal):
    def sound(self):
        print(f"{self.name} meows") # print petname + it meows 

# Create method for Dog and Cat class but it will call sound method of respective class not same way for both classes 

d= Dog("Buddy")
d.sound()   # Output: Buddy barks
c= Cat("Whiskers")
c.sound()   # Output: Whiskers meows        
```

<h1>Encapsulation>
<h5>Encapsulation means wrapping data and methods together and protecting data from direct access.(dataprotection)<h5>

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # Private variable (protected)

    def deposit(self, amount):
        self.__balance += amount

    def show_balance(self):
        print("Balance:", self.__balance)

# Create object
acc = BankAccount(5000)

acc.deposit(1000)        # Correct way to change balance
acc.show_balance()

# acc.__balance  (cannot access directly)
```

<h1>Abstraction<h1>
<h5>Abstraction means hiding how things work and showing only what is necessary.<h5>


```python
from abc import ABC, abstractmethod

# Abstract parent class
class Vehicle(ABC):

    @abstractmethod
    def move(self):
        pass    # Only method name, no code


# Child class 1
class Car(Vehicle):
    def move(self):
        print("Car moves on road")


# Child class 2
class Bike(Vehicle):
    def move(self):
        print("Bike moves on road")


# Create objects
c = Car()
b = Bike()

# Same method, different implementations
c.move()
b.move()
```

```python
from abc import ABC, abstractmethod   # for abstraction


# -------- ABSTRACTION --------
class Animal(ABC):

    def __init__(self, name):
        self.name = name          # public variable
        self.__age = 2            # private variable (ENCAPSULATION)

    # getter (access private data)
    def get_age(self):
        return self.__age

    # setter (change private data)
    def set_age(self, age):
        self.__age = age

    @abstractmethod
    def sound(self):              # must be implemented by child
        pass


# -------- INHERITANCE --------
class Dog(Animal):

    def sound(self):
        return "Dog barks"


class Cat(Animal):

    def sound(self):
        return "Cat meows"


# -------- POLYMORPHISM --------
def animal_sound(animal):
    print(animal.sound())


# -------- OBJECTS --------
dog = Dog("Tommy")
cat = Cat("Kitty")

# -------- ENCAPSULATION --------
dog.set_age(5)
cat.set_age(3)

print(dog.name, "age is", dog.get_age())
print(cat.name, "age is", cat.get_age())

# -------- POLYMORPHISM IN ACTION --------
animal_sound(dog)
animal_sound(cat)
```
