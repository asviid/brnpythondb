```python
# using variables to store data
a = 10  #integer variable
name="John" #string variable
price=99.99 #float variable
is_valid=True   #boolean variable
# to printing variables data by using print statement
print(a)
print(name) 
print(price)
print(is_valid)
#  i can do like this to print all variables in single print statement
print(a, name, price, is_valid)
# to check the data type of variables
print(type(a))
print(type(name))
print(type(price))
print(type(is_valid))
```

```python
#Type casting in python
x = "10" # string variable
y = int(x) #casting string to integer

print(y + 5) # Output: 15
```

```python
# input from user and output the same
# in the same way we can take multiple inputs from user and perform operations as needed and convert data types accordingly


name= input("Enter your name: ")  # input function to take input from user default data type is string
print("Hello, " + name + "! Welcome to Python programming.")

age= int(input("Enter your age: ")) # converting string input to integer using int() 
print("You are " + str(age) + " years old.")  # converting integer to string using str() for concatenation

height= float(input("Enter your height in meters: ")) # converting string input to float using float()
print("Your height is " + str(height) + " meters.")  # converting float to string using str() for concatenation
```

```python
#arthmetic operations
num1= int(input("Enter first number: "))
num2= int(input("Enter second number: "))
print("Addition: ", num1 + num2) # Addition
print("Subtraction: ", num1 - num2) # Subtraction
print("Multiplication: ", num1 * num2) # Multiplication     
print("Division: ", num1 / num2) # Division just give float value quotient
print("Modulus: ", num1 % num2) # Modulus gives remainder as output integer data type
print("Exponentiation: ", num1 ** num2) # Exponentiation
print("Floor Division: ", num1 // num2) # Floor Division  gives integer value quotient
# using comments in python 
```

```python
#comparison operators in python

a = 5
b = 10

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

```python
#logical operators in python    
x = True
y = False

print(x and y)
print(x or y)
print(not x)
```

```python
x = 10
x += 5  # which is equivalent to x = x + 5
print(x)

x -= 3  # which is equivalent to x = x - 3
print(x)

x *= 2  # which is equivalent to x = x * 2
print(x)
```

```python
# bitwise operators in python
a = 5  # 0101
b = 3  # 0011

print(a & b)  # AND
print(a | b)  # OR
print(a ^ b)  # XOR
print(~a)     # NOT
print(a << 1) # Left shift
print(a >> 1) # Right shift
```

```python
# Membership Operators
list1 = [1, 2, 3, 4, 5]
print(3 in list1)  # True
print(6 in list1)  # False
print(3 not in list1)  # False
print(6 not in list1)  # True

# Identity Operators
a = 10
b = 10
print(a is b)      # True
print(a is not b)  # False  

#ternary operator in python
age = 20
status = "Adult" if age >= 18 else "Minor"
print(status)  # Output: Adult
```

```python
#codintional statements in python
num = int(input("Enter a number: "))    
if num > 0:
    print("The number is positive.")    
elif num == 0:
    print("The number is zero.")
else:
    print("The number is negative.")
    
```

```python
#for loop in python
for i in range(1, 6):
    print(i)
```

```python
#while loop in python
count = 1
while count <= 5:
    print(count)
    count += 1
    
```

```python
#break and continue statements in loops
for i in range(5):
    if i == 3:
        break
    print(i)
for j in range(5):
    if j == 3:
        continue
    print(j)
```

```python
#functions in python
def square(x):
    return x * x


square(5)# function call

print(square(5)) # Output: 25 
```
